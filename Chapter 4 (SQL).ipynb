### 4.1

#### a) Using the same sqlite database we were working with in this chapter, use SQL to make a DataFrame that's summarizes at bats, hits and homeruns at the player-game level. Make it only AL East Batters. It should have the following collums: date, team, opp, batter, n_atbats, hit, homerun. Rename hit and homerun to just H and HR respectively.

```python
import pandas as pd
from os import path
import sqlite3
```

```python
conn = sqlite3.connect('baseball.sqlite')
```

```python
df = pd.read_sql(
    """
    SELECT date, player_game.team, player_game.opp, batter, n_atbats, hit AS H, homerun AS HR
    FROM player_game, game, team
    WHERE
        player_game.g_id = game.g_id AND
        player_game.team = team.teamID AND
        team.lgID = 'AL' AND
        team.divID = 'E'
    """, conn)

df
```

#### b) Now modify your query so that instead of batter you have first_name and last_name from the player table. Use abbreviation too (t for the team table, pg for player_game, etc)

```python
df = pd.read_sql(
    """
    SELECT date, pg.team, pg.opp, p.first_name, p.last_name, n_atbats, hit AS H, homerun AS HR
    FROM player_game AS pg, game AS g, team AS t, player AS p
    WHERE
        pg.g_id = g.g_id AND
        pg.team = t.teamID AND
        p.id = pg.batter_id AND
        t.lgID = 'AL' AND
        t.divID = 'E'
    """, conn)

df
```

### 4.2 

#### Using the same sqlite database, make a DataFrame that's the game table, with two new columsn: 'home_full' and 'away_full' that have the full names of the teams (e.g. New York Yankees, Milwaukee Brewers)

```python
df = pd.read_sql(
    """
    SELECT g.*, th.name as home_full, ta.name as away_full
    FROM game AS g, team AS th, team AS ta
    WHERE
        g.home_team = th.teamID AND
        g.away_team = ta.teamID
    """, conn)

df.head()
```
